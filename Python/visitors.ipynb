```python
import requests
import pandas as pd
from lxml import etree
from bs4 import BeautifulSoup
import calendar
import csv
import json
```

```python
months = [i for i in range(1, 9)]
days = [calendar.monthrange(2020, m)[1] for m in months]


months
days
```

```python
def r_get(url, opt=1):
    header = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.79 Safari/537.36"}

    r = requests.get(url, headers=header)
    r.encoding = 'utf-8'

    if opt == 0:
        return BeautifulSoup(r.text, "lxml")
    else:
        return etree.HTML(r.text)
```

```python
def filling(num):
    return f"0{str(num)}" if num < 10 else num
```

```python
data = {}

for i, month in enumerate(months):
    for day in range(1, days[i]+1):
        url = f"https://www.immd.gov.hk/hkt/stat_2020{filling(month)}{filling(day)}.html"

        tree = r_get(url)

        data[f"2020/{filling(month)}/{filling(day)}"] = [int(td.text.replace(",", "").strip()) for td in tree.xpath("/html/body/div/table/tbody/tr[21]/td[@class='hRight'][4]")]
```

```python
data
```

```python
df=pd.DataFrame.from_dict(data, orient='index').dropna()
df = df.reset_index().rename(columns={0:'Entries','index':'Date'})
df.to_csv('Entry.csv',index=0) 
df
```
